```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler,LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('Student_Performance(new).csv')
df.head()
```

```python
X = df.drop(columns='Performance Index')
y=df['Performance Index']
```

```python
num_cols=X.select_dtypes(include='number').columns
obj_cols=X.select_dtypes(include='object').columns
```

```python
sns.boxplot(X[num_cols])
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
scaling = MinMaxScaler()
xtrain[num_cols] = scaling.fit_transform(xtrain[num_cols])
xtrain
```

```python
df.corr(numeric_only=True)
```

```python
xtrain[obj_cols].nunique()
```

```python
xtrain[obj_cols]
```

```python
#label_encoder = LabelEncoder()
#xtrain['Extracurricular Activities'] = label_encoder.fit_transform(xtrain['Extracurricular Activities'])
#xtrain
```

```python
encoder = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
values = encoder.fit_transform(xtrain[obj_cols])
cols = encoder.get_feature_names_out()
```

```python
xtrain[cols] = values
xtrain.drop(columns=['Extracurricular Activities'],inplace=True)
xtrain
```

```python
#updating xtest before model training
xtest[num_cols] = scaling.transform(xtest[num_cols])
xtest_values = encoder.transform(xtest[obj_cols])
xtest[cols] = xtest_values
xtest.drop(columns=['Extracurricular Activities'],inplace=True)
xtest
#now that xtest is updated, we can train the model
```

```python
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
#Evaluation
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
y_pred_test=model.predict(xtest)
y_pred_test
```

```python
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error,mean_absolute_error
```

```python
r2 = r2_score(ytest, y_pred_test)
print("R2 Score:", r2)
```

```python
mae = mean_absolute_error(ytest, y_pred_test)
print("MAE:", mae)
```

```python
import numpy as np
rmse = np.sqrt(mean_squared_error(ytest, y_pred_test))
print("RMSE:", rmse)
```
